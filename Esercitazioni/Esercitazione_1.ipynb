```python
from datetime import datetime
# calculate using now the today string in format "2023-01-01" 
today = datetime.now().strftime("%Y-%m-%d")
```

```python
import requests
import pandas as pd

# Milano 
lat, lon = 45.4642, 9.1900

# Open-Meteo – storico orario
url = "https://archive-api.open-meteo.com/v1/archive"

params = {
    "latitude": lat,
    "longitude": lon,
    "hourly": ["temperature_2m", "relative_humidity_2m", "windspeed_10m"],
    "start_date": "2025-01-01",
    "end_date": today,
    "timezone": "UTC"
}

r = requests.get(url, params=params, timeout=20)
# DataFrame temperatura
hourly = r.json()["hourly"]
historical = pd.DataFrame(hourly)
historical["time"] = pd.to_datetime(historical["time"])
```

### Dati forecast


```python
# Open-Meteo – storico orario
url = "https://api.open-meteo.com/v1/forecast"

params = {
    "latitude": lat,
    "longitude": lon,
    "hourly": ["temperature_2m", "relative_humidity_2m", "windspeed_10m"],
    "timezone": "UTC"}

r = requests.get(url, params=params, timeout=20)
# DataFrame temperatura
hourly = r.json()["hourly"]
forecast = pd.DataFrame(hourly)
forecast["time"] = pd.to_datetime(forecast["time"])
```
