## 2. Come funziona Jupyter Notebook?

Jupyter Notebook si basa sulla creazione di documenti chiamati **notebook**, che possono essere eseguiti su un server locale o remoto. Un notebook Ã¨ composto da una serie di **celle** (*cells*), che possono contenere codice, testo formattato con la sintassi *Markdown*, immagini e altro.

Le celle di codice possono essere eseguite interattivamente, permettendo di vedere immediatamente i risultati delle operazioni effettuate. I risultati possono essere visualizzati all'interno del notebook, sotto forma di grafici, tabelle e altro.

## 3. Come utilizzare Jupyter Notebook?

Per creare un nuovo notebook dalla schermata iniziale (che mostra il filesystem), clicca sul pulsante **"New"** in alto a destra e seleziona **"Python 3 (ipykernel)"**.

Una volta creato il notebook, puoi scrivere codice all'interno delle celle di codice. Per eseguire una cella di codice, clicca sulla cella e premi il pulsante **"Run"** in alto oppure premi **Shift + Invio** sulla tastiera.

Per aggiungere testo formattato con la sintassi *Markdown*, seleziona una cella e cambia il tipo di cella in **"Markdown"** dal menu a tendina in alto. Puoi utilizzare diversi comandi *Markdown* per creare titoli, elenchi, link, immagini e altro.

> ðŸ’¡ **Suggerimento:** Per salvare il notebook, utilizza la combinazione di tasti **Ctrl + S**.

## 7. Shortcut utili

Ecco alcuni **shortcut** utili che possono rendere piÃ¹ efficiente l'utilizzo di Jupyter Notebook:

- **ModalitÃ  Comando** (premi `Esc` per attivarla):
    - `A`: Inserisci una nuova cella **sopra** la cella selezionata.
    - `B`: Inserisci una nuova cella **sotto** la cella selezionata.
    - `M`: Cambia il tipo di cella in **Markdown**.
    - `Y`: Cambia il tipo di cella in **Code**.
    - `D` `D` (premi due volte): Elimina la cella selezionata.
- **ModalitÃ  Edit** (premi `Enter` per attivarla):
    - `Ctrl + S`: Salva il notebook.
    - `Shift + Enter`: Esegui la cella corrente e seleziona quella successiva.
    - `Ctrl + Shift + -`: Dividi la cella corrente in due nel punto del cursore.

> ðŸ’¡ **Suggerimento:** Per una lista completa degli shortcut, premi **`H`** in modalitÃ  comando.
